```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("health_armaan.csv", parse_dates=["date"]).sort_values("date").reset_index(drop=True)
```

```python
monthly_rhr = (
    df.groupby(["year","month"])["RestingHeartRate"]
      .mean()
      .reset_index()
      .sort_values(["year","month"])
)

plt.figure(figsize=(12,4))
plt.bar(monthly_rhr.index, monthly_rhr["RestingHeartRate"])
plt.xticks(
    ticks=monthly_rhr.index,
    labels=[f"{y}-{m:02d}" for y,m in zip(monthly_rhr["year"], monthly_rhr["month"])],
    rotation=45, ha="right"
)
plt.title("Average Resting Heart Rate per Month")
plt.xlabel("Month"); plt.ylabel("Resting HR (bpm)")
plt.tight_layout(); plt.show()

monthly_rhr
```

```python
monthly_hrv = (
    df.groupby(["year","month"])["HeartRateVariabilitySDNN"]
      .mean()
      .reset_index()
      .sort_values(["year","month"])
)

plt.figure(figsize=(12,4))
plt.bar(monthly_hrv.index, monthly_hrv["HeartRateVariabilitySDNN"])
plt.xticks(
    ticks=monthly_hrv.index,
    labels=[f"{y}-{m:02d}" for y,m in zip(monthly_hrv["year"], monthly_hrv["month"])],
    rotation=45, ha="right"
)
plt.title("Average HRV (SDNN) per Month")
plt.xlabel("Month"); plt.ylabel("HRV (ms)")
plt.tight_layout(); plt.show()

monthly_hrv
```

```python
monthly_walk_hr = (
    df.groupby(["year","month"])["WalkingHeartRateAverage"]
      .mean()
      .reset_index()
      .sort_values(["year","month"])
)

plt.figure(figsize=(12,4))
plt.bar(monthly_walk_hr.index, monthly_walk_hr["WalkingHeartRateAverage"])
plt.xticks(
    ticks=monthly_walk_hr.index,
    labels=[f"{y}-{m:02d}" for y,m in zip(monthly_walk_hr["year"], monthly_walk_hr["month"])],
    rotation=45, ha="right"
)
plt.title("Average Walking HR per Month")
plt.xlabel("Month"); plt.ylabel("Walking HR (bpm)")
plt.tight_layout(); plt.show()

monthly_walk_hr
```

```python
monthly_fitness = (
    df.groupby(["year","month"])["HRFitnessIndex"]
      .mean()
      .reset_index()
      .sort_values(["year","month"])
)

plt.figure(figsize=(12,4))
plt.bar(monthly_fitness.index, monthly_fitness["HRFitnessIndex"])
plt.xticks(
    ticks=monthly_fitness.index,
    labels=[f"{y}-{m:02d}" for y,m in zip(monthly_fitness["year"], monthly_fitness["month"])],
    rotation=45, ha="right"
)
plt.title("Average HR Fitness Index per Month")
plt.xlabel("Month"); plt.ylabel("HR Fitness Index")
plt.tight_layout(); plt.show()

monthly_fitness
```
